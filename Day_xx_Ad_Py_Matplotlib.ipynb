```python
plt.title("sales Indication")
plt.xlabel('sales per year')
plt.ylabel('revenue per year')
plt.plot(x,y)
plt.show()
```

```python
# Addition of Z axis 
```

```python
x = [1,2,3]
y = [2,5,7]
z = [10,3,1]
plt.plot(x,y)
plt.plot(x,z)
plt.title('this is a test graph')
plt.xlabel('x-axis')
plt.ylabel('y and z axis')
plt.legend(["this is y","this is z"])
plt.show()


```

```python
# 2nd CSV file
# Import 2nd file  as "countries.csv"
```

```python
data = pd.read_csv('countries.csv')
data
```

```python
# Req: Population Grwoth Comparision Between 10 diffrent countries.

# USA & China

#step:1:Getting the USA data from main data file
#step:2:Getting the china data from main data file.
```

```python
us = data[data.country == 'United States']
us
```

```python
india = data[data.country == 'India']
india
```

```python
plt.plot(us.year, us.population/10**6)
plt.plot(india.year, india.population/10**6)
plt.xlabel('year')
plt.ylabel('population')
plt.title('population growth comparison')
plt.legend(['united states','India'])
plt.show()
```
